```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import signal
import math
```

```python
t = np.linspace(0, 1000, 1000, endpoint=True)
sig = np.linspace(0, 1000, 1000, endpoint=True)
sig[0:500]  = np.cos(2 * np.pi * 5 * t[0:500])
sig[500:1000]  = np.cos(2 * np.pi * 20 * t[500:1000])
# sig  = np.cos(2 * np.pi * 7 * t)
# sig  = signal.gausspulse(t - 0.4, fc=2)
plt.plot(t, sig)
plt.show()
```

```python
a_start = 1
a_stop = 100
a_step = .5
widths = np.arange(a_start, a_stop, a_step)

w_start = a_start * len(sig) / a_stop
```

```python
cwtmatr = signal.cwt(sig, signal.ricker, widths)

plt.plot(t, sig)
plt.show()

plt.imshow(cwtmatr, extent=[0, 1000, a_stop*4, 0], cmap='PRGn', aspect='auto',
               vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
# plt.imshow(cwtmatr, cmap='PRGn', aspect='auto',
#                vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.colorbar()
plt.show()
```
