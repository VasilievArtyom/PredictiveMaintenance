```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import signal
```

```python
t = np.linspace(-1, 1, 601, endpoint=True)
sig  = np.cos(2 * np.pi * 7 * t) + signal.gausspulse(t - 0.4, fc=3)
# sig  = np.cos(2 * np.pi * 7 * t)
# sig  = signal.gausspulse(t - 0.4, fc=2)
plt.plot(t, sig)
plt.show()
w_start = 1
w_stop = 31
w_step = 0.2
widths = np.arange(w_start, w_stop, w_step)
cwtmatr = np.abs(signal.cwt(sig, signal.ricker, widths))
# plt.imshow(cwtmatr, extent=[-1, 1, w_start, w_stop], cmap='PRGn', aspect='auto',
#                vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.imshow(cwtmatr, cmap='PRGn', aspect='auto',
               vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.colorbar()
plt.show()
```
